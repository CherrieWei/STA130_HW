````python
import numpy as np
import plotly.graph_objs as go
from scipy.stats import norm
import plotly.io as pio

# Set the default renderer to 'notebook'
pio.renderers.default = 'notebook'

# Parameters for the linear regression model
beta_0 = 2  # Intercept
beta_1 = 0.5  # Slope
sigma = 1  # Standard deviation of errors

# Generate sample data for x
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # 100 evenly spaced values from 0 to 10

# Generate the deterministic part of y (the true line)
y_true = beta_0 + beta_1 * x

# Generate random errors from a normal distribution with mean 0 and standard deviation sigma
errors = np.random.normal(0, sigma, size=x.size)

# Observed y values with errors added
y_observed = y_true + errors

# Initialize the Plotly figure
fig = go.Figure()

# Add scatter plot for observed data
fig.add_trace(go.Scatter(x=x, y=y_observed, mode='markers', name='Observed Data', marker=dict(color='red', opacity=0.5)))

# Add line for the true model (deterministic part)
fig.add_trace(go.Scatter(x=x, y=y_true, mode='lines', name='True Line', line=dict(color='blue')))

# Plotting the normal distribution of y at a specific x value for illustration
x_point = 5  # Choose a specific x value to illustrate the distribution of y|x
y_mean_at_x = beta_0 + beta_1 * x_point
y_values = np.linspace(y_mean_at_x - 3*sigma, y_mean_at_x + 3*sigma, 100)
pdf = norm.pdf(y_values, loc=y_mean_at_x, scale=sigma)

# Plot the normal distribution vertically at x_point
fig.add_trace(go.Scatter(
    x=[x_point] * len(y_values),
    y=y_values,
    mode='lines',
    line=dict(color='green', dash='dash'),
    name='Normal Distribution at x=5'
))

# Adjust layout for the plot
fig.update_layout(
    title="Simple Linear Regression Model with Normal Error Distribution",
    xaxis_title="x",
    yaxis_title="y",
    legend=dict(x=0.8, y=0.95)
)

# Show the figure
fig.show(renderer='png')
````

````python
import pandas as pd
import numpy as np

# Simulating data (example code)
beta_0 = 2  # Intercept
beta_1 = 0.5  # Slope
sigma = 1  # Standard deviation of errors
np.random.seed(0)  # For reproducibility

# Generate sample data for x
x = np.linspace(0, 10, 100)  # 100 evenly spaced values from 0 to 10

# Generate the deterministic part of Y (the true line)
y_true = beta_0 + beta_1 * x

# Generate random errors from a normal distribution
errors = np.random.normal(0, sigma, size=x.size)

# Observed Y values with errors added
Y = y_true + errors

# Create a DataFrame
df = pd.DataFrame({
    'x': x,  # x values
    'Y': Y   # Observed Y values
})

# Display the DataFrame
print(df.head())
````

````python
import statsmodels.formula.api as smf  # the library provides classes and functions for estimating and testing statistical model
import plotly.express as px  # this is a ploting library

#this line specifies a linear regression model using the ordinary least squares OLS method
model_data_specification = smf.ols("Y~x", data=df) 

#this line fits the specified linear regression model to the data in the DataFrame
fitted_model = model_data_specification.fit() 

fitted_model.summary()  # generates a comprehensive summary table of the fitted regression model
fitted_model.summary().tables[1]  # accesses a second table in the summary output, which typically contains information specifically about the regression coefficients
fitted_model.params  # returns the estimated coefficients of the regression model, providing a Pandas Series where the index corresponds to the variable names and the values are the estimated coefficients
fitted_model.params.values  # retrieves just the numerical values of the estimated coefficients without the index
fitted_model.rsquared  # provides the R-squared value of the fitted model

# what two things does this add onto the figure?
# 1. creates a legend entry labeled "Data" that appears in the figures legend with all points sharing the same color
# 2. visually indicates the relationship between Y and x based on a linear regression analysis
df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='Y', color='Data', 
                 trendline='ols', title='Y vs. x')

# This is essentially what above `trendline='ols'` does
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

fig.show(renderer='png')
````

````python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import plotly.express as px

# Simulating data (example code)
beta_0 = 2  # Intercept
beta_1 = 0.5  # Slope
sigma = 1  # Standard deviation of errors
np.random.seed(0)  # For reproducibility

# Generate sample data for x
x = np.linspace(0, 10, 100)  # 100 evenly spaced values from 0 to 10

# Generate the deterministic part of Y (the true line)
y_true = beta_0 + beta_1 * x

# Generate random errors from a normal distribution
errors = np.random.normal(0, sigma, size=x.size)

# Observed Y values with errors added
Y = y_true + errors

# Create a DataFrame
df = pd.DataFrame({
    'x': x,  # x values
    'Y': Y   # Observed Y values
})

# Fit the OLS model
model_data_specification = smf.ols("Y ~ x", data=df) 
fitted_model = model_data_specification.fit() 

# Create the scatter plot with trendline
df['Data'] = 'Data'  # Hack to add data to legend
fig = px.scatter(df, x='x', y='Y', color='Data', 
                 trendline='ols', title='Y vs. x')

# Add the OLS trendline
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="OLS Trendline")

# Add the true line to the figure
fig.add_scatter(x=x, y=y_true, 
                line=dict(color='green', dash='dash'), name="True Line")

# Show the figure
fig.show(renderer='png')
````

Differences:
1. The OLS trendline represents the best fit line obtained from ordinary least squares method, which takes the difference between the observed and predicted values into account, while the true line represents the actual relationship between the predictor and outcome variable without the error term and randomness.
2. The OLS trendline works to minimize the sum of squared differences between the observed values and the fitted values, therefore it might adjust its slope and intercept based on the distribution of the observed data. This can potentially causes the OLS trendline to appear differently from the true mean.

A line of best fit is chosen for the fitted model based on the observed data using OLS method. The OLS method attempts to find the values of 0 and 1 that minimizes the difference between the observed and predicted outcome values or the residuals. More specifically, it minimizes the sum of the squared residuals. It requires squaring because this amplifies the influence of larger residuals, meaning that the outliers have a disproportionately larger effect on the fit of the model. Squaring also ensures the values are non-negative because distance cannot be negative. The squared residuals also create a continuous function that is differentiable.

> _**R-squared** is the "the proportion of variation in (outcome) $Y$ explained by the model ($\hat y_i$)" and is defined as_
>
> _$R^2 = 1 - \frac{\sum_{i=1}^n(Y_i-\hat y)^2}{\sum_{i=1}^n(Y_i-\bar Y)^2}$_
>
> _The visuzation provided in the previous problem can be used to consider $(Y_i-\bar Y)^2$ as the squared distance of the $Y_i$ to their sample average $\bar Y$ as opposed to the squared **residuals** $(Y_i-\hat y)^2$ which is the squared distance of the $Y_i$ to their fitted (predicted) values $Y_i$._

The first expression is a formulation of the coefficient of determination, which indicates how well the regression model explains the variation in the dependent variable. `Y-fitted_model.fittedvalues` represents the residuals, and squaring and summing these values give the total squared error. The fraction of total squared error over the total sum of squares represents the proportion of total variation that is not explained by the model, hence 1 - ((Y-fitted_model.fittedvalues)**2).sum()/((Y-Y.mean())**2).sum() gives the proportion of variation that is explained by the model.

`fitted_model.rsquared` is interpreted as a measure of the accuracy of the model because it reflects the proportion of variance in the outcome variable that is explained by the model's predictors. The higher the r-squared value, the higher the proportion of variation can be explained by the model, meaning a higher accuracy of the model.

`np.corrcoef(Y, fitted_model.fittedvalues)[0, 1]**2` captures the squared correlation coefficient between the observed dependent variable and the predicted values obtained from the regression model, meaning the accuracy of the regression model. `np.corrcoef(Y, x)[0, 1]**2` calculates the squared correlation coefficient between the observed dependent variable and the independent variable, meaning the strength of the linear relationship between the variables.

````python
import numpy as np
import pandas as pd
import seaborn as sns

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Create a 'kind' column based on a median waiting time cutoff to classify "short" and "long" groups
median_waiting = old_faithful['waiting'].median()
old_faithful['kind'] = np.where(old_faithful['waiting'] < median_waiting, 'short', 'long')

# Calculate the observed difference in mean duration between the two groups
observed_diff = old_faithful.groupby('kind')['duration'].mean().diff().iloc[-1]

# Permutation test
num_permutations = 10000
diffs = np.zeros(num_permutations)

for i in range(num_permutations):
    # Shuffle the 'kind' labels and assign to a new column
    old_faithful['kind_shuffled'] = old_faithful['kind'].sample(n=len(old_faithful), replace=False).values
    
    # Calculate the difference in mean duration between the shuffled groups
    mean_diff = old_faithful.groupby('kind_shuffled')['duration'].mean().diff().iloc[-1]
    diffs[i] = mean_diff

# Calculate p-value
p_value = np.mean(np.abs(diffs) >= np.abs(observed_diff))

print(f"Observed Difference: {observed_diff}")
print(f"P-value: {p_value}")
````

````python
import numpy as np
import pandas as pd

# Assuming you have two groups of data: group_1 and group_2
# Example data for demonstration:
np.random.seed(42)  # For reproducibility
group_1 = np.random.normal(loc=50, scale=10, size=100)
group_2 = np.random.normal(loc=55, scale=10, size=100)

# Function to calculate the mean difference between two groups
def bootstrap_mean_difference(group_1, group_2, n_iterations=10000):
    bootstrapped_mean_differences = []
    
    for _ in range(n_iterations):
        # Resample with replacement from both groups
        sample_1 = np.random.choice(group_1, size=len(group_1), replace=True)
        sample_2 = np.random.choice(group_2, size=len(group_2), replace=True)
        
        # Calculate the mean difference for the bootstrap sample
        mean_diff = np.mean(sample_1) - np.mean(sample_2)
        bootstrapped_mean_differences.append(mean_diff)
    
    # Convert to a numpy array for easy quantile computation
    bootstrapped_mean_differences = np.array(bootstrapped_mean_differences)
    
    # Calculate the 95% confidence interval (2.5th and 97.5th percentiles)
    ci_lower, ci_upper = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])
    
    return ci_lower, ci_upper

# Get the 95% confidence interval for the mean difference
ci_lower, ci_upper = bootstrap_mean_difference(group_1, group_2)
print(f"95% Confidence Interval for the Mean Difference: ({ci_lower}, {ci_upper})")
````

Bootstrapping assumes that the observed data is a good representation of the population and that resampling with replacement allows for approximating the distribution of statistics, even without making strong assumptions about the populations distribution. The permutation test is focused on rearranging the data to test the null hypothesis that there is no effect, while bootstrapping focuses on estimating the sampling distribution of a statistic by resampling from the observed data. In bootstrapping, we resample within a single dataset to create new samples and estimate the variability of a statistic. In permutation testing, we shuffle the group assignments to test whether the observed difference is likely to have arisen by chance under the null hypothesis of no effect.

Compare and contrast:
Bootstrapping leverages the actual dataset and re-samples within it to create variability estimates. It reflects how the observed sample's inherent variability might affect the slope estimate, giving insight into the confidence interval without presuming a zero slope. Simulation under the Null Hypothesis relies on synthetic data generated under the assumption of no association. This method provides a null distribution for testing how likely the observed slope is under this no-relationship assumption, helping calculate a simulated p-value.

The OLS regression is computationally simpler than either bootstrapping or simulation but assumes that the residuals are normally distributed and that the true model structure is linear. The standard p-value and confidence interval provided by OLS are based on these assumptions, potentially less flexible than bootstrapping for small samples or skewed data.

Both bootstrapping and OLS provide confidence intervals that allow checking if zero falls within the interval, which would suggest no association. The p-value from the null hypothesis simulation may differ from the OLS p-value due to differences in assumptions. If the null hypothesis distribution from simulations resembles the OLS distribution, the p-values should align closely. Discrepancies may arise if the OLS assumptions are not fully met by the data.

All three approaches ultimately aim to assess the slope coefficient in the linear regression model (duration ~ waiting). Each approach provides insights into the strength and direction of the relationship between waiting time and duration by examining the estimated slope. Both bootstrapping and OLS regression provide confidence intervals for the slope. These intervals allow comparison with zero to infer whether a statistically significant linear relationship exists between waiting and duration. While the null hypothesis simulation doesnt directly yield a confidence interval, it allows for p-value interpretation that aligns with whether zero is a plausible slope. All three methods ultimately test the null hypothesis that there is no association between waiting time and duration.
